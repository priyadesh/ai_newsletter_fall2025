```python
# Import required libraries
import requests
import json
from datetime import datetime
from pprint import pprint

# API Base URL
BASE_URL = "http://localhost:8080/api/v1"

# Helper function to make API calls pretty
def call_api(endpoint, method="GET", data=None):
    """Make an API call and return the response"""
    url = f"{BASE_URL}{endpoint}"
    
    if method == "GET":
        response = requests.get(url)
    elif method == "POST":
        response = requests.post(url, json=data)
    elif method == "DELETE":
        response = requests.delete(url)
    
    return response.json()

print("‚úÖ Setup complete!")
```

```python
# Check API health
health = call_api("/health")

print(f"üìä Health Status: {health['data']['status']}")
print(f"üìù Version: {health['data']['version']}")
print(f"üîß Mode: {health['data']['mode']}")
print("\nüîå Services:")
for service, status in health['data']['services'].items():
    icon = "‚úÖ" if status == "connected" else "‚ö†Ô∏è"
    print(f"  {icon} {service}: {status}")

print("\n‚öôÔ∏è  Configuration:")
print(f"  ‚Ä¢ Max articles: {health['data']['configuration']['max_articles']}")
print(f"  ‚Ä¢ Cache TTL: {health['data']['configuration']['cache_ttl_minutes']} minutes")
print(f"  ‚Ä¢ News query: {health['data']['configuration']['news_query']}")
```

```python
# Fetch the newsletter
print("üì∞ Fetching newsletter...")
newsletter = call_api("/newsletter")

if newsletter['success']:
    data = newsletter['data']
    
    # Display editorial
    print("\n" + "="*80)
    print("üìù EDITORIAL")
    print("="*80)
    print(f"\nTitle: {data['editorial']['title']}")
    print(f"Author: {data['editorial']['author']}")
    print(f"\n{data['editorial']['content'][:300]}...")
    
    # Display article summaries
    print("\n" + "="*80)
    print(f"üìö NEWS ARTICLES ({data['total_articles']} total)")
    print("="*80)
    
    for i, summary in enumerate(data['summaries'][:3], 1):  # Show first 3
        print(f"\n{i}. {summary['catchy_title']}")
        print(f"   Source: {summary['original_article']['source']}")
        print(f"   Relevance: {summary['relevance_score']*100:.0f}%")
        print(f"   Summary: {summary['summary'][:150]}...")
    
    print(f"\n... and {data['total_articles'] - 3} more articles")
else:
    print(f"‚ùå Error: {newsletter['message']}")
```

```python
# Get the first article's ID
article_id = data['summaries'][0]['id']

# Fetch the article
print(f"üìÑ Fetching article: {article_id}")
article_response = call_api(f"/newsletter/{article_id}")

if article_response['success']:
    article = article_response['data']
    
    print("\n" + "="*80)
    print("üìÑ ARTICLE DETAILS")
    print("="*80)
    print(f"\nTitle: {article['title']}")
    print(f"Source: {article['source']}")
    print(f"URL: {article['url']}")
    print(f"Relevance: {article['relevance_score']*100:.0f}%")
    
    print(f"\nüìù Summary:\n{article['summary']}")
    
    print(f"\nüîë Key Points:")
    for i, point in enumerate(article['key_points'], 1):
        print(f"  {i}. {point}")
else:
    print(f"‚ùå Error: {article_response['message']}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data for visualization
sources = [s['original_article']['source'] for s in data['summaries']]
relevance_scores = [s['relevance_score'] for s in data['summaries']]
titles = [s['catchy_title'][:30] + '...' for s in data['summaries']]

# Create visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Relevance scores
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(relevance_scores)))
bars = ax1.barh(range(len(titles)), relevance_scores, color=colors)
ax1.set_yticks(range(len(titles)))
ax1.set_yticklabels(titles, fontsize=8)
ax1.set_xlabel('Relevance Score')
ax1.set_title('Article Relevance Scores')
ax1.set_xlim(0, 1)
ax1.grid(axis='x', alpha=0.3)

# Plot 2: Sources distribution
from collections import Counter
source_counts = Counter(sources)
ax2.pie(source_counts.values(), labels=source_counts.keys(), autopct='%1.1f%%', startangle=90)
ax2.set_title('News Sources Distribution')

plt.tight_layout()
plt.show()

# Print statistics
print("\nüìä Newsletter Statistics:")
print(f"  ‚Ä¢ Total articles: {len(data['summaries'])}")
print(f"  ‚Ä¢ Average relevance: {np.mean(relevance_scores):.2f}")
print(f"  ‚Ä¢ Unique sources: {len(source_counts)}")
print(f"  ‚Ä¢ Top source: {source_counts.most_common(1)[0][0]}")
```
